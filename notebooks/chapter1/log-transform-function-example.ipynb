```python
# Author: Guillaume Lemaitre <guillaume.lemaitre@inria.fr>
# License: BSD 3 clause

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import RidgeCV
from sklearn.compose import TransformedTargetRegressor
from sklearn.metrics import median_absolute_error, r2_score
from sklearn.utils.fixes import parse_version

density_param = {"density": True}
```

```python
# generate synthetic uniform random data
X, y = make_regression(n_samples=10000, noise=100, random_state=0)

# transform the target, y, into an exponential distribution
y = np.expm1((y + abs(y.min())) / 200)

# apply the log transform function on y
y_trans = np.log1p(y)
```

```python
f, (ax0, ax1) = plt.subplots(1, 2)

ax0.hist(y, bins=100, **density_param)
ax0.set_xlim([0, 2000])
ax0.set_ylabel("Probability")
ax0.set_xlabel("Target")
ax0.set_title("Exponential distribution")

ax1.hist(y_trans, bins=100, **density_param)
ax1.set_ylabel("Probability")
ax1.set_xlabel("Target")
ax1.set_title("Normal distribution")

f.tight_layout(rect=[0.0, 0.0, 0.95, 0.95])
```
