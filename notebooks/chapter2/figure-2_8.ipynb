```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="darkgrid")
df = sns.load_dataset("iris")
df.rename(columns = {'sepal_length':'sales_hour', 'sepal_width':'sales_year'}, inplace = True)


sns.histplot(data=df, x="sales_hour", color="skyblue", label="Training", kde=True)
sns.histplot(data=df, x="sales_year", color="red", label="Serving", kde=True)

plt.legend() 
plt.show()
```
